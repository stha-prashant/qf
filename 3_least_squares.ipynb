```python
import numpy as np
def simpleLS(x, y, a=None):
    if a is None:
        a  = (np.sum(x*y))/np.sum(x**2)
    R = np.sum((a*x-y)**2)
    return a, R
```

```python
x = np.array([0.23, 0.88, 0.21, 0.92, 0.49, 0.62, 0.77, 0.52, 0.30, 0.19]).reshape(-1, 1)
y = np.array([0.19, 0.96, 0.33, 0.80, 0.46, 0.45, 0.67, 0.32, 0.38, 0.37]).reshape(-1, 1)


a, R = simpleLS(x,y)
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x, a*x)
plt.title(f"R={R}")
plt.show()
```

```python
def affineLS(x,y, a=None, b=None):
    if a is None:
        a = (np.sum((x-np.mean(x))*(y-np.mean(y))))/np.sum((x-np.mean(x))**2)
    if b is None:
        b = np.mean(y) - a*np.mean(x)
    R = np.sum((a*x + b - y)**2)
    return a, b, R
```

```python
a, b, R = affineLS(x,y)
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x, a*x+b)
plt.title(f"R={R}")
plt.show()
```
